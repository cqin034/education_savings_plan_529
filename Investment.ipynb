```python
#equity/bond

acctBalance = float(1000)
#account balance

EQUITY_RETURN_COL = 6
DATE_COL = 1
BOND_RETURN_COL = 22
startMonth = 1548
PERCENT_EQUITY = .1
#% of investment that is diretcted to equities (instead of bonds)
INVEST_FREQ = 12
#how many times a year money is added to the balance (12, 4, 2, 1)
YEARLY_INVEST_BUDGET = 1200
#how much invested additionally

equityRateOfReturn = worksheet.col_values(EQUITY_RETURN_COL)
bondRateOfReturn = worksheet.col_values(BOND_RETURN_COL)
date = worksheet.col_values(DATE_COL)


#remove headers
for i in range(8):
  equityRateOfReturn.pop(0)
for i in range(8):
  bondRateOfReturn.pop(0)
for i in range(8):
    date.pop(0)
#reformat date for vega editor (javascript)
for i in range(1831):
  if len(date[i]) != 7:
    date[i] = date[i].replace('.','-')
    date[i] = date[i] + '0'
  else:
    date[i] = date[i].replace('.','-')

month = startMonth
#reassign so the variable can change

for i in range(228):
  if i != 0 and (i % (12/INVEST_FREQ) == 0):
    acctBalance += YEARLY_INVEST_BUDGET/INVEST_FREQ

  eROR = float(equityRateOfReturn[month])
  #eROR = equity rate of return
  bROR = float(bondRateOfReturn[month])
  #bROR = bond rate of return
  amtEquity = acctBalance * PERCENT_EQUITY
  #amount invested into equities
  amtBond = acctBalance - amtEquity
  #amount invested into bonds


  acctBalance = (amtEquity * (1 + eROR)) + (amtBond * (1 + bROR))

  print("{\"percent equity\": \"" + str(PERCENT_EQUITY) + "\", \"date\": \"" + date[month] + "-01T00:00:00\", \"balance\": " + str(acctBalance) + "},")

  month += 1
```
